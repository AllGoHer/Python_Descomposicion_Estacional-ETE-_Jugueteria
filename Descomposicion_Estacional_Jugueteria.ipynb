```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
juguetes= pd.read_csv('C:\\Users\\ALLAN\\Desktop\\Business Analitics con Python y ChatGPT\\juguetes.csv', index_col='mes')
```

```python
juguetes.index= pd.to_datetime(juguetes.index)
```

```python
juguetes.head()
```

```python
juguetes.plot()
plt.show()
```

```python
#Descomposicion en modelo aditivo
descomposicion=seasonal_decompose(juguetes['ventas'], model='additive')
```

```python
#Visualizar Tendencias
descomposicion.trend.plot()
plt.title('Tendecia de Ventas Juguetes')
plt.xlabel('Mes')
plt.ylabel('Miles de USD')
plt.show()
```

```python
#Visualizar la estacionalidad
descomposicion.seasonal.plot()
plt.title('Estacionalidad de Ventas Juguetes')
plt.xlabel('Mes')
plt.ylabel('Miles de USD')
plt.show() 
```

```python
#Visualizar los residuos
descomposicion.resid.plot()
plt.title('Residuos de Ventas Juguetes')
plt.xlabel('Mes')
plt.ylabel('Miles de USD')
plt.show() 
```

```python
#Visualizar todos los componentes juntos
fig=descomposicion.plot()
fig.set_size_inches(10,8)
fig.suptitle('Descomposicion de Ventas de Juguetes', fontsize=14)
plt.show()
```
